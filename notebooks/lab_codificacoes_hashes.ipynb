
# 🔐 Laboratório: Codificações e Hashes em Segurança da Informação

Este laboratório permite que você insira um texto e veja como ele é representado em diferentes tipos de codificações, bem como os hashes mais comuns utilizados em segurança da informação. 

**Nota:** Hashes **não são** codificações – são funções unidirecionais utilizadas para verificação de integridade, autenticação e segurança.


```

text = input("Digite um texto para codificação e hashing: ")

import base64
import urllib.parse
import binascii
import hashlib
import quopri

print("\n--- Codificações ---")
print(f"Base64: {base64.b64encode(text.encode()).decode()}")
print(f"URL Encoding: {urllib.parse.quote(text)}")
print(f"Hexadecimal (binário para hex): {text.encode().hex()}")
print(f"ASCII Encoding (lista de códigos): {[ord(c) for c in text]}")
print(f"Hexadecimal Encoding (por caractere): {' '.join(hex(ord(c)) for c in text)}")
print(f"UTF-8 Encoding: {list(text.encode('utf-8'))}")
print(f"Quoted Printable: {quopri.encodestring(text.encode()).decode()}")

print("\n--- Hashes (não são codificações) ---")
print(f"MD5: {hashlib.md5(text.encode()).hexdigest()}")
print(f"SHA1: {hashlib.sha1(text.encode()).hexdigest()}")
print(f"SHA256: {hashlib.sha256(text.encode()).hexdigest()}")

print("\n✔️ Fim das codificações e hashes.")
```
